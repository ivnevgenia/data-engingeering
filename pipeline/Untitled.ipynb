```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata_2025', con=engine))
```

```python
df.head(n=0).to_sql(name='green_tripdata_2025', con=engine, if_exists='replace')
```

```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet')
from tqdm.auto import tqdm
# Добавляем прогресс-бар
for _ in tqdm([df], desc='Saving to SQL'):
    df.to_sql(name='green_tripdata_2025', con=engine, if_exists='append')
```
